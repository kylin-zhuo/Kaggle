Feature engineering

```python
def feature_engineering(df):
    # map the name to its length: there might be correaltion in the name's length
    df['NameLength'] = df['Name'].apply(len)
    # map the gender
    df['Sex'] = df['Sex'].map({'female': 1, 'male': 0}).astype(int)
    # map the age
    age_avg = df['Age'].mean()
    age_std = df['Age'].std()
    age_null_count = df['Age'].isnull().sum()
    age_null_random_list = np.random.randint(age_avg - age_std, age_avg + age_std, size=age_null_count)
    df['Age'][np.isnan(df['Age'])] = age_null_random_list
    df['Age'] = df['Age'].astype(int)
    df.loc[ df['Age'] <= 16, 'Age'] = 0
    df.loc[(df['Age'] > 16) & (df['Age'] <= 32), 'Age'] = 1
    df.loc[(df['Age'] > 32) & (df['Age'] <= 48), 'Age'] = 2
    df.loc[(df['Age'] > 48) & (df['Age'] <= 64), 'Age'] = 3
    df.loc[ df['Age'] > 64, 'Age'] = 4
    # family size
    df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
    df['IsAlone'] = 0
    df.loc[df['FamilySize'] == 1, 'IsAlone'] = 1
    # fill in the NaN of embarked 
    df['Embarked'] = df['Embarked'].fillna('S')
    df['Embarked'] = df['Embarked'].map( {'S': 0, 'C': 1, 'Q': 2} ).astype(int)
    # map hasCabin
    df['HasCabin'] = df['Cabin'].apply(lambda x: 0 if type(x) == float else 1)
    # map fare
    df.loc[ df['Fare'] <= 7.91, 'Fare'] = 0
    df.loc[(df['Fare'] > 7.91) & (df['Fare'] <= 14.454), 'Fare'] = 1
    df.loc[(df['Fare'] > 14.454) & (df['Fare'] <= 31), 'Fare']   = 2
    df.loc[ df['Fare'] > 31, 'Fare'] = 3
    df['Fare'] = df['Fare'].astype(int)
    # finalize
    drop_cols = ['PassengerId', 'Name', 'Ticket', 'Cabin', 'SibSp']
    df.drop(drop_cols, axis = 1, inplace=True)
```

```python
df = pd.read_csv('train.csv')
```

```python
feature_engineering(df)
```
